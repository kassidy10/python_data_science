```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig = plt.figure()
ax = plt.axes()
plt.show
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)

fig = plt.figure()
ax = plt.axes()

ax.plot(x,y)
plt.show()
```

```python
x = np.linspace(0,10,1000)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function sin(x)')
plt.show
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = 'k')
plt.plot(x, np.cos(x), color = 'r', linestyle = '--')
plt.show()
```

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), 'k', label = 'sin(x)')
plt.plot(x, np.cos(x), 'r--', label = 'cos(x)')
plt.legend()
plt.show()
```

Scatter Plot

```python
presidents_df = pd.read_csv('https://sololearn.com/uploads/files/president_heights_party.csv', index_col='name')
```

```python
plt.scatter(presidents_df['height'], presidents_df['age'])
plt.show()
```

```python
plt.scatter(presidents_df['height'], presidents_df['age'], marker = '<', color = 'b')
plt.xlabel('height');
plt.ylabel('age')
plt.title('U.S presidents')
plt.show()
```

```python
presidents_df.plot(kind = 'scatter',
                   x = 'height',
                   y = 'age',
                   title = 'U.S presidents')
plt.show()
```

Histogram

```python
presidents_df['height'].plot(kind = 'hist',
                             title = 'height',
                             bins = 5)
plt.show()
```

```python
plt.hist(presidents_df['height'], bins = 5)
plt.show()
```

Boxplot

```python
presidents_df['height'].describe()
```

```python
plt.style.use('classic')
presidents_df.boxplot(column='height')
```

BarPlot

```python
party_cnt = presidents_df['party'].value_counts()

plt.style.use('ggplot')
party_cnt.plot(kind = 'bar')
plt.show()
```
